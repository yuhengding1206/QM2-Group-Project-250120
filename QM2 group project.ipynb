```python
# Comparison of the pirate attacks in Aden Gulf & Guinea Gulf & Southeast Asia from 1993 to 2020
import pandas as pd
import requests
from io import StringIO
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline

# GitHub URL (do not modify)
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'

# Use requests to fetch the CSV file content
response = requests.get(url)
data = StringIO(response.text)

# Read the CSV file
df = pd.read_csv(data)

# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Define the date range
start_date = '1995-01-01'
end_date = '2020-12-31'

# Filter data for Aden Gulf
filtered_data_Aden = df[df['nearest_country'].isin(['YEM', 'SOM', 'ERI', 'DJI'])]
filtered_data_Aden = filtered_data_Aden[(filtered_data_Aden['date'] >= start_date) & (filtered_data_Aden['date'] <= end_date)]
quarterly_counts_Aden = filtered_data_Aden.resample('Q', on='date').size()

# Filter data for Gulf of Guinea
filtered_data_Guinea = df[df['nearest_country'].isin(['NGA', 'LBR', 'GHA', 'BEN', 'STP', 'CIV', 'GIN', 'TGO', 'GNQ', 'GAB', 'CMR'])]
filtered_data_Guinea = filtered_data_Guinea[(filtered_data_Guinea['date'] >= start_date) & (filtered_data_Guinea['date'] <= end_date)]
quarterly_counts_Guinea = filtered_data_Guinea.resample('Q', on='date').size()

# Filter data for Asia
filtered_data_Asia = df[df['nearest_country'].isin(['PHL', 'VNM', 'KHM', 'THA', 'MYS', 'IDN', 'SGP'])]
filtered_data_Asia = filtered_data_Asia[(filtered_data_Asia['date'] >= start_date) & (filtered_data_Asia['date'] <= end_date)]
quarterly_counts_Asia = filtered_data_Asia.resample('Q', on='date').size()

# Convert indices to quarterly format
quarterly_counts_Aden.index = quarterly_counts_Aden.index.to_period('Q').strftime('%Y Q%q')
quarterly_counts_Guinea.index = quarterly_counts_Guinea.index.to_period('Q').strftime('%Y Q%q')
quarterly_counts_Asia.index = quarterly_counts_Asia.index.to_period('Q').strftime('%Y Q%q')

# Ensure quarterly_counts_Aden has all quarters from 1995 Q1 to 2020 Q4
all_quarters = pd.period_range(start=start_date, end=end_date, freq='Q').strftime('%Y Q%q').tolist()
quarterly_counts_Aden = quarterly_counts_Aden.reindex(all_quarters, fill_value=0)
quarterly_counts_Guinea = quarterly_counts_Guinea.reindex(all_quarters, fill_value=0)
quarterly_counts_Asia = quarterly_counts_Asia.reindex(all_quarters, fill_value=0)


# Function to create smooth curves
def smooth_curve(x, y):
    x_new = np.linspace(0, len(x) - 1, 300)  # 300 points for smoothness
    spline = make_interp_spline(np.arange(len(x)), y, k=3)  # Cubic spline
    y_new = spline(x_new)
    return x_new, y_new

# Smooth the data
x_Aden, y_Aden = smooth_curve(quarterly_counts_Aden.index, quarterly_counts_Aden.values)
x_Guinea, y_Guinea = smooth_curve(quarterly_counts_Guinea.index, quarterly_counts_Guinea.values)
x_Asia, y_Asia = smooth_curve(quarterly_counts_Asia.index, quarterly_counts_Asia.values)

plt.figure(figsize=(12, 6))
plt.plot(x_Aden, y_Aden, color='b', label='Aden Gulf')
plt.plot(x_Guinea, y_Guinea, color='g', label='Guinea Gulf')
plt.plot(x_Asia, y_Asia, color='r', label='Southeast Asia')

plt.title('Figure 4:Number of Pirate Attacks in Aden Gulf & Guinea Gulf & Southeast Asia from 1995 Q1 to 2020 Q4')
plt.xlabel('Time (Quarter)')
plt.ylabel('Number of Pirate Attacks')

# Set x-axis ticks to display only Q1 of each year
xticks = [f"{year} Q1" for year in range(1995, 2021)]
xtick_positions = [quarterly_counts_Aden.index.get_loc(xtick) for xtick in xticks if xtick in quarterly_counts_Aden.index]

plt.xticks(xtick_positions, xticks, rotation=45)  # Rotate x-axis labels for better readability

# Add a label for the vertical line
plt.axvline(x=quarterly_counts_Aden.index.get_loc('2008 Q4'), color='Purple', linestyle='--',label='Intervention Starts')

# Adjust legend position to ensure the label is included
plt.legend(loc='upper right')

plt.grid()
plt.tight_layout()  # Adjust layout to avoid label overlap
plt.show()
```

```python
# Pallel Analysis for Aden Gulf and Guinea Gulf:

import pandas as pd
import requests
from io import StringIO
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols

# Load data
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'
response = requests.get(url)
data = StringIO(response.text)
df = pd.read_csv(data)

# Ensure 'date' column is datetime
df['date'] = pd.to_datetime(df['date'])

# Filter data for the relevant time period
start_date = '1995-01-01'
end_date = '2008-12-31'  # Change to before the policy implementation
df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Define treatment and control groups
# Treatment group: Aden Gulf (Yemen and Somalia)
# Control group: Southeast Asia (specific countries)
df['treatment'] = np.where(df['nearest_country'].isin(['YEM', 'SOM', 'ERI', 'DJI']), 1,
                            np.where(df['nearest_country'].isin(['NGA', 'LBR', 'GHA', 'BEN', 'STP', 'CIV', 'GIN', 'TGO', 'GNQ', 'GAB', 'CMR']), 0, np.nan))

# Remove rows where treatment is NaN (not in either group)
df = df.dropna(subset=['treatment'])

# Create quarter variable
df['quarter'] = df['date'].dt.to_period('Q')  # Create period-based quarter
agg_df = df.groupby(['quarter', 'treatment']).size().reset_index(name='num_attacks')

# Convert 'quarter' to string for plotting
agg_df['quarter'] = agg_df['quarter'].astype(str)

# **Fix for incomplete data: Ensure all quarters from 1995 Q1 to 2008 Q4 are present**
all_quarters = pd.period_range(start='1995Q1', end='2008Q4', freq='Q').astype(str).tolist()
idx = pd.MultiIndex.from_product([all_quarters, [0, 1]], names=['quarter', 'treatment'])
agg_df = agg_df.set_index(['quarter', 'treatment']).reindex(idx, fill_value=0).reset_index()

# Visualize the trends for treatment and control groups
plt.figure(figsize=(12, 6))
sns.lineplot(data=agg_df, x='quarter', y='num_attacks', hue='treatment', marker='o')
plt.title(' Figure 5: Trends in Pirate Attacks in Aden Gulf & Guinea Gulf Before Operation Atalanta (1995 Q1 -2008 Q4)')
plt.ylabel('Number of Pirate Attacks')
plt.xlabel('Time (Quarter)')

# Add vertical dashed line for intervention start at 2008 Q4
plt.axvline(x='2008Q4', color='red', linestyle='--', linewidth=2, label='Intervention Starts')

# Change x-tick labels to show corresponding year and "Q1" for all Q1s
unique_quarters = agg_df['quarter'].unique()
plt.xticks(ticks=np.arange(len(unique_quarters)),
           labels=[f"{quarter.split('Q')[0]} Q1" if 'Q1' in quarter else '' for quarter in unique_quarters], rotation=45)

# Add grid
plt.grid(True)  # Ensure grid is enabled
plt.tight_layout()
plt.show()

# Run regression to test for parallel trends
agg_df['time'] = np.arange(len(agg_df))  # Create a time variable for the regression and name it 'time'

# Run the regression for the pre-policy period
pre_did_model = ols('num_attacks ~ treatment + time + treatment*time', data=agg_df).fit()
print(pre_did_model.summary())  # Added parenthesis here

# Conclusion based on the regression results
if pre_did_model.pvalues['treatment:time'] < 0.05:
    print("Therefore: The assumptions for a parallel analysis are not met.")
else:
    print("Therefore: The assumptions for a parallel analysis are met.")
```

```python
# Differences in differences analysis for Aden Gulf and Guinea Gulf: (Completed)

import pandas as pd
import requests
from io import StringIO
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols

# Load data
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'
response = requests.get(url)
data = StringIO(response.text)
df = pd.read_csv(data)

# Ensure 'date' column is datetime
df['date'] = pd.to_datetime(df['date'])

# Filter data for the relevant time period
start_date = '1995-01-01'
end_date = '2020-12-31'  # Updated to include data until the end of 2020
df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Define treatment and control groups
# Treatment group: Aden Gulf (Yemen and Somalia)
# Control group: Southeast Asia (specific countries)
df['treatment'] = np.where(df['nearest_country'].isin(['YEM', 'SOM', 'ERI', 'DJI']), 1,
                            np.where(df['nearest_country'].isin(['NGA', 'LBR', 'GHA', 'BEN', 'STP', 'CIV', 'GIN', 'TGO', 'GNQ', 'GAB', 'CMR']), 0, np.nan))

# Remove rows where treatment is NaN (not in either group)
df = df.dropna(subset=['treatment'])

# Create quarter variable
df['quarter'] = df['date'].dt.to_period('Q')  # Create period-based quarter
agg_df = df.groupby(['quarter', 'treatment']).size().reset_index(name='num_attacks')

# Convert 'quarter' to string for plotting
agg_df['quarter'] = agg_df['quarter'].astype(str)

# Create 'post' variable:
agg_df['post'] = np.where((agg_df['quarter'] >= '2011Q1') & (agg_df['quarter'] <= '2020Q4'), 1, 0)
agg_df['post_treatment'] = agg_df['post'] * agg_df['treatment']

# Run DID regression
did_model = ols('num_attacks ~ post + treatment + post_treatment', data=agg_df).fit()
print(did_model.summary())

# Visualize boxplot for pre and post periods by treatment group
# The 'post' column was created in agg_df, not df.
# Use agg_df to create the 'category' column.
agg_df['category'] = agg_df['treatment'].astype(str) + agg_df['post'].astype(str)
# Use agg_df for the boxplot data
sns.boxplot(x='category', y='num_attacks', hue='treatment', data=agg_df).set_xticklabels(
    ['Pre x Control', 'Pre x Treatment', 'Post x Control', 'Post x Treatment']
)
plt.title('Number of Pirate Attacks by Treatment and Post Periods')
plt.xlabel('')
plt.ylabel('Number of Pirate Attacks')
plt.legend(title='Region')
plt.tight_layout()
plt.show()

# The regression was using df instead of agg_df, which has 'post', 'treatment', 'post_treatment', and 'num_attacks'
df_model = ols('num_attacks ~ post + treatment + post_treatment', data=agg_df).fit()
print(df_model.summary())
```

```python
import pandas as pd
import requests
from io import StringIO

# GitHub URL (unchanged)
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'

# Use requests to fetch the CSV file content
response = requests.get(url)
data = StringIO(response.text)

# Load the CSV file into a DataFrame
df = pd.read_csv(data)

# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Define the list of countries in the Gulf of Guinea region (with corresponding ISO codes)
guinea_gulf_countries = ['NGA', 'BEN', 'TGO', 'GHA', 'CIV', 'CMR']

# Filter the data for the Gulf of Guinea region
filtered_data_Guinea_Gulf = df[df['nearest_country'].isin(guinea_gulf_countries)]

# Ensure the data is within the range of 2000 to 2020
start_date = '2000-01-01'
end_date = '2020-12-31'
filtered_data_Guinea_Gulf = filtered_data_Guinea_Gulf[(filtered_data_Guinea_Gulf['date'] >= start_date) & (filtered_data_Guinea_Gulf['date'] <= end_date)]

# Calculate the number of pirate attacks per year
annual_counts = filtered_data_Guinea_Gulf.set_index('date').resample('Y').size()

# Filter the data for years 2000 to 2020
annual_counts_filtered = annual_counts[(annual_counts.index.year >= 2000) & (annual_counts.index.year <= 2020)]

# Output the number of pirate attacks in the Gulf of Guinea from 2000 to 2020
print("Pirate Attacks in Gulf of Guinea from 2000 to 2020:")
total_attacks = 0
for year in range(2000, 2021):
    count = annual_counts_filtered.get(pd.Timestamp(f"{year}-12-31"), 0)  # Get the data for each year
    total_attacks += count
    print(f"{year}: {count} attacks")

print(f"\nTotal attacks from 2000 to 2020: {total_attacks} attacks")
```

```python
import pandas as pd
import requests
from io import StringIO
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf

# GitHub CSV data URL
url = 'https://raw.githubusercontent.com/yuhengding1206/QM2-2024/main/Per%20Attack%20Statistics.csv'

# Load the CSV data from GitHub
response = requests.get(url)
df = pd.read_csv(StringIO(response.content.decode('utf-8')))

# Filter for Aden Gulf and Guinea Gulf regions
df = df[df['Region'].isin(['Aden_Gulf', 'Guinea_Gulf'])]

# Convert the Year column to numeric
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Create a variable indicating whether the year is post-2009
df['post'] = np.where(df['Year'] >= 2010, 1, 0)

# Create treatment and control group variables
df['treatment'] = np.where(df['Region'] == 'Aden_Gulf', 1,
                           np.where(df['Region'] == 'Guinea_Gulf', 0, np.nan))

# Create an interaction term for post-treatment
df['post_treatment'] = df['post'] * df['treatment']

# Create a time variable
df['time'] = df['Year']

# Filter data for 2000-2009
df_pre_2009 = df[(df['Year'] >= 2000) & (df['Year'] < 2010)]

# Plot trends for each region
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_pre_2009, x='Year', y='Per_Attack', hue='Region', marker='o')
plt.title('Attack Count Trend (2000-2009) for Aden Gulf and Guinea Gulf')
plt.xlabel('Year')
plt.ylabel('Number of Attacks (Per Attack)')
plt.axvline(x=2009, color='red', linestyle='--', label='Atalanta Operation')
plt.legend()
plt.grid(True, linestyle='--', linewidth=0.7)
plt.show()

# Fit the regression model for pre-2010 data
pre_df_model = smf.ols('Per_Attack ~ treatment + time + treatment*time', data=df[df['Year'] < 2010]).fit()
print(pre_df_model.summary())

# Conclusion based on the regression results
if pre_df_model.pvalues['treatment:time'] < 0.05:
    print("Therefore: The assumptions for a parallel analysis are not met.")
else:
    print("Therefore: The assumptions for a parallel analysis are met.")
```
